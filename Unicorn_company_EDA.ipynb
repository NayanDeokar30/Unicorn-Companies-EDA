```python
unicorn["Date Joined"] = pd.to_datetime(unicorn["Date Joined"])
unicorn.head()
```

```python
unicorn["month"] = unicorn["Date Joined"].dt.month
unicorn.head()
```

```python
unicorn['month_txt'] = unicorn['Date Joined'].dt.month_name().str.slice(stop=3)
```

```python
unicorn_by_month = unicorn.groupby(['month','month_txt']).sum().sort_values('month', ascending = True).head(12).reset_index()
unicorn_by_month
```

```python
plt.bar(x=unicorn_by_month['month_txt'],height=unicorn_by_month['Year Founded'])

plt.xlabel("months")
plt.ylabel("Year founded")
plt.title("No. of unicorn founded by months")
plt.show()
```

```python
# to change the column name
# since column name with spaces cannot be used so we put underscore inplace of every space
unicorn.rename(columns = {"Year Founded" : "Year_Founded"}, inplace = True)
unicorn.rename(columns = {"Date Joined" : "Date_Joined"}, inplace = True)
unicorn.rename(columns = {"Select Investors" : "Select_Investors"}, inplace = True)
unicorn.head()
```

```python
unicorn.Year_Founded.value_counts()
```

```python
unicorn.Year_Founded.value_counts()[:20].rename_axis("unique_Values").reset_index(name = "counts").style.background_gradient()
```

```python
unicorn["week"] = unicorn.Date_Joined.dt.isocalendar().week
unicorn["weekday"] = unicorn.Date_Joined.dt.day_name()
unicorn.head()
```

```python
unicorn[["weekday", "Year_Founded"]].groupby(["weekday"]).mean()
```

```python
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
# this is the box plot with outliers
g = sns.boxplot ( data = unicorn,
                 x = "weekday",
                 y = "Year_Founded",
                 order = weekday_order,
                 showfliers = True
                );
g.set_title("Year founded per weekday");
```

```python
# this is the box plot without outliers
g = sns.boxplot(data = unicorn,
               x = "weekday",
               y = "Year_Founded",
               order = weekday_order,
               showfliers = False);
g.set_title("Year founded per weekday")
```
